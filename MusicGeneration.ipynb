```python
import torch
from transformers import MusicgenForConditionalGeneration, AutoProcessor

device = "cuda" if torch.cuda.is_available() else "cpu"
processor = AutoProcessor.from_pretrained("facebook/musicgen-small")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-small").to(device)
```

```python
prompt = ["lofi chill beat with soft piano and rain"]

input = processor(text=prompt, padding=True, return_tensors="pt").to(device)
```

```python
audio_values = model.generate(**input,max_new_tokens=512)

import scipy.io.wavfile
scipy.io.wavfile.write("output.wav",rate=32000,data=audio_values[0, 0].cpu().numpy())
```
